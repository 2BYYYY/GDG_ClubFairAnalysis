# The dataset of how many pictures are taken during the clubfair
**data** = transformed data in the image_names.csv file

```python
import pandas as pd

# CSV to pandas dataframe
data = pd.read_csv("image_names.csv")

# Split the name into date format
split_date = data["pictures"].str.split().str[1]
date = pd.to_datetime(split_date, format="%Y-%m-%d").dt.date
data["date"] = date

# Split the name into time format
split_pics = data["pictures"].str.split().str[2].str.split('.').str[0]
time = pd.to_datetime(split_pics, format='%H%M%S').dt.time
data["time"] = time
```

# DAY 1 of pictures
**dayOne** = All pictures during day 1

**countDayOne** = The count of pictures during day 1

**dayOneFirstPic** = Day 1 First Pic

**dayOneLastPic** = Day 1 Last Pic 
## AM (9 - 12)  PM (12 - 6)
**amDayOne** = All pictures dring AM

**countAmDayOne** = The count of pictures during AM

**pmDayOne** = All pictures during PM

**countPmDayOne** = The count of pictures during PM

```python
# data[[]] is not the same as data[data[]] 
# data[data[]] needs to have a condition like dayOne (true,false)
# pd.concat()
# is called to combine all the DataFrames stored in amDayOne_list and pmDayOne_list into two single DataFrames (amDayOne and pmDayOne).

dayOne = data[
    (data["pictures"].str.contains("2024-09-09")) |  
    (data["pictures"].str.contains("2024-09-10"))
]
dayOneFirstPic = dayOne.head(1)
dayOneLastPic = dayOne.tail(1)
countDayOne = dayOne["pictures"].count()

amDayOne_list = []
pmDayOne_list = []

for i in range(18):
    hour = str(i).zfill(2)     
    
    if i < 12:
        amDayOne_ = dayOne[
            (dayOne["pictures"].str.contains(f"2024-09-09 {hour}")) | 
            (dayOne["pictures"].str.contains(f"2024-09-10 {hour}"))
        ]
        amDayOne_list.append(amDayOne_)
    else:
        pmDayOne_ = dayOne[
            (dayOne["pictures"].str.contains(f"2024-09-09 {hour}")) |
            (dayOne["pictures"].str.contains(f"2024-09-10 {hour}"))
        ]
        pmDayOne_list.append(pmDayOne_)

amDayOne = pd.concat(amDayOne_list)
countAmDayOne = amDayOne["pictures"].count()
pmDayOne = pd.concat(pmDayOne_list)
countPmDayOne = pmDayOne["pictures"].count()
```

# DAY 2 of pictures
**dayTwo** = All pictures during day 2

**countDayTwo** = The count of pictures during day 2

**dayTwoFirstPic** = Day 2 First Pic

**dayTwoLastPic** = Day 2 Last Pic

# DAY 3 of pictures
**dayThree** = All pictures during day 3

**countDayThree** = The count of pictures during day 3

**dayThreeFirstPic** = Day 3 First Pic

**dayThreeLastPic** = Day 3 Last Pic
## AM (9 - 12)  PM (12 - 6)

**countPmDayThree** = The count of pictures during PM

```python
dayThree = data[(data["pictures"].str.contains("2024-09-12"))]
countDayThree = dayThree["pictures"].count()

dayThree = data[
    (data["pictures"].str.contains("2024-09-12"))
]
dayThreeFirstPic = dayThree.head(1)
dayThreeLastPic = dayThree.tail(1)
countdayThree = dayThree["pictures"].count()

amdayThree_list = []
pmdayThree_list = []

for i in range(18):
    hourThree = str(i).zfill(2)     
    
    if i < 12:
        amdayThree_ = dayThree[
            (dayThree["pictures"].str.contains(f"2024-09-12 {hourThree}")) 
        ]
        amdayThree_list.append(amdayThree_)
    else:
        pmdayThree_ = dayThree[
            (dayThree["pictures"].str.contains(f"2024-09-12 {hourThree}"))
        ]
        pmdayThree_list.append(pmdayThree_)

amDayThree = pd.concat(amdayThree_list)
countAmDayThree = amDayThree["pictures"].count()
pmDayThree = pd.concat(pmdayThree_list)
countPmDayThree = pmDayThree["pictures"].count()
```

# Pictures taken per hour

# Conclusion the pictures peeks at (11am-12pm or 3pm-4pm)

```python
import matplotlib.pyplot as plt
import seaborn as sns

# new DataFrame of sorted time and sorted time in hours
sorted_time = data["time"].sort_values(ascending=True)
sorted_time = pd.DataFrame(sorted_time)
sorted_time["hour"] = pd.to_datetime(data["time"], format="%H:%M:%S").dt.hour
count_Hour = sorted_time["hour"].value_counts().sort_index()


# time in hours in Day 1
dayOne_time = dayOne["time"]
dayOne_time = pd.DataFrame(dayOne_time)
hourDayOne = pd.to_datetime(dayOne_time["time"], format ="%H:%M:%S" ).dt.hour
countDayOne_time = hourDayOne.value_counts().sort_index()

# time in hours in Day 2
dayTwo_time = dayTwo["time"]
dayTwo_time = pd.DataFrame(dayTwo_time)
hourDayTwo = pd.to_datetime(dayTwo_time["time"], format ="%H:%M:%S" ).dt.hour
countDayTwo_time = hourDayTwo.value_counts().sort_index()

# time in hours in Day 3
dayThree_time = dayThree["time"]
dayThree_time = pd.DataFrame(dayThree_time)
hourDayThree = pd.to_datetime(dayThree_time["time"], format ="%H:%M:%S" ).dt.hour
countDayThree_time = hourDayThree.value_counts().sort_index()

# Plot using pandas plot function Day 1
countDayOne_time.plot(kind='line', marker='o', title="Hour Counts for Day 1")
plt.xlabel("Hour of the Day")
plt.ylabel("Count of Time Occurrences")
plt.show()

countDayTwo_time.plot(kind='line', marker='o', title="Hour Counts for Day 2")
plt.xlabel("Hour of the Day")
plt.ylabel("Count of Time Occurrences")
plt.show()

countDayThree_time.plot(kind='line', marker='o', title="Hour Counts for Day 3")
plt.xlabel("Hour of the Day")
plt.ylabel("Count of Time Occurrences")
plt.show()

count_Hour.plot(kind='line', marker='o', title="Hour Counts")
plt.xlabel("Hour of the Day")
plt.ylabel("Count of Time Occurrences")
plt.show()
```

# The number of pictures taken each day from the club fair

```python
days = ['DAY 1', 'DAY 2', 'DAY 3']
count = [countDayOne, countDayTwo, countDayThree]

barOne = plt.bar(days, count)
plt.bar_label(barOne, label_type='center')
plt.show()
```

# The day when most pictures were taken

```python
explode = (0, 0.1, 0)
plt.pie(count, explode, labels=days, autopct='%.2f%%', shadow=True, startangle=90)
plt.show()
```

# Club Fair AM vs PM

```python
amClubFairCount = (countAmDayOne + countAmDayTwo + countAmDayThree)
pmClubFairCount = (countPmDayOne + countPmDayTwo + countPmDayThree)

explode = [0, 0.1]

amVSpmHeader = [f"AM = {amClubFairCount}", f"PM = {pmClubFairCount}"]
amVSpmValue = [amClubFairCount,  pmClubFairCount]
plt.pie(amVSpmValue, explode, labels=amVSpmHeader, autopct='%.2f%%', shadow=True)
plt.show()
```

# Pictures taken each day during AM and PM

```python
import numpy as np

AmPm = {
    'Am': np.array([countAmDayOne, countAmDayTwo, countAmDayThree]),
    'Pm': np.array([countPmDayOne, countPmDayTwo, countPmDayThree]),
}
bottom = np.zeros(3)

for time, ampm in AmPm.items():
    p = plt.bar(days, ampm, label=time, bottom=bottom)
    bottom += ampm
    plt.bar_label(p, label_type="center")
    plt.legend()

plt.show()
```
